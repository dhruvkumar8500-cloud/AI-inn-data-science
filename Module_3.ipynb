```python
import pandas as pd
import matplotlib.pyplot as plt
dic = {"Marks" : [34,56,78]}

df=pd.DataFrame(dic)
df.index = ["Dhruv", "Amit", "Mayank"]
print(df)

plt.plot(df["Marks"],marker="o",linestyle=":",linewidth=2,color="g")
plt.ylabel("Marks",color='r',fontsize=10)
plt.xlabel("Subject")
plt.title("Marks of students")
plt.grid(True)
plt.show()
```

## **Pie Chart**

```python
sub =["Physics", "Chemistry", "Maths"]
marks = [34,45,67]
plt.pie(marks,labels=sub,autopct="%.2f%%",startangle=60,explode=[0.05,0.07,0.08],shadow=True)
plt.show()
```

```python
#Data points with the help of loop
x = [2,3,4,5,6]
y = [6,7,8,5,4]
for i in range(len(x)):
  print("Data Point", i+1, ":", x[i], y[i])
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Data Points using Loop")
plt.show()
```

```python
#Two Line Single Graph
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [5,15,25,35,45]

for i in range(len(x)):
  print("Line 1 Data Point:", x[i], y1[i])
  print("Line 2 Data Point:", x[i], y2[i])

plt.plot(x, y1, marker='o', label='Line 1')
plt.plot(x, y2, marker='o', label='Line 2')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Lines in a Single Graph")
plt.legend()
plt.show()
```

```python
#Add legend in graph
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [5,15,25,35,45]

plt.plot(x, y1, marker='o', label='Line 1')
plt.plot(x, y2, marker='o', label='Line 2')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Lines in One Graph")
plt.legend()     # This adds the legend
plt.show()
```

```python
plt.figure(figsize=(8, 6))

#First subplot
plt.subplot(2, 1, 1) #(rows, columns, position)
plt.plot(x, y1, marker='o', label='Line Chart 1')
plt.title("First Line Chart")
plt.legend()

#Second subplot
plt.subplot(2, 1, 2)
plt.plot(x,y2, marker='s', label='Line Chart 2')
plt.title("Second Line Chart")
plt.legend()

plt.tight_layout()
plt.show()
```

## **Bar Chart**

```python
import matplotlib.pyplot as plt

name = ["Mayank", "Dhruv", "Shivam", "Dev", "Aayush"]
marks = [34,56,78,67,45]

plt.figure(figsize=(10, 5))
plt.bar(name, marks, color=['red', 'yellow', 'pink', 'blue', 'green'], width=0.7, label='Marks')
plt.xlabel("Name")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.legend()
plt.grid(axis='y')
for index,value in enumerate(marks):
  plt.text(index, value + 1, str(value), ha='center') # Adjust position for better visibility
plt.show()
```
